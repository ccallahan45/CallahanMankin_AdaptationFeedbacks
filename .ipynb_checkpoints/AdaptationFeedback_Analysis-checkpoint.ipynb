#### Functions

Function for loading SSP data and calculating quantities for Kaya Identity

```python
def load_ssp_data(scenario,subset,mdl,years):
    # load data for desired variables, interpolate to yearly timestep
    
    var = ["Population","Emissions|CO2","GDP|PPP","Final Energy"]
    vname = ["population","emissions","gdp","energy"]

    # scenario name, load
    scenario_nm = scenario+"-"+subset
    data_in = pd.read_csv(loc_ssp_iam+"SSP_IAM_V2_201811.csv")
    data = data_in.loc[(data_in.SCENARIO==scenario_nm)&(data_in.REGION=="World")&(data_in.MODEL==mdl),:]
    
    # build dataframe
    df_orig = pd.DataFrame({"year":years,"population":np.full(len(years),np.nan),
                  "gdp":np.full(len(years),np.nan),"energy":np.full(len(years),np.nan),
                  "emissions":np.full(len(years),np.nan)})
    
    # load each variable
    for v in var:
        vn = vname[var.index(v)]
        data_v = data.loc[data.VARIABLE.values==v,:]
        yrs_v = data_v.iloc[:,7:].columns.values.astype(int) # index 7 because 2020
        data_v_vals = data_v.iloc[:,7:].values[0] # index 7 because 2020
        df_ind = [x in yrs_v for x in df_orig.year.values]
        df_orig.loc[df_ind,vn] = data_v_vals
    
    # convert US 2005 to US 2020
    gdp_deflator = pd.read_csv(loc_deflator+"API_NY.GDP.DEFL.ZS_DS2_en_excel_v2_4353530.csv")
    usa_deflator = gdp_deflator.loc[gdp_deflator["Country Code"]=="USA",:]
    deflator_2020_2005 = usa_deflator.loc[:,"2020"].values[0]/usa_deflator.loc[:,"2005"].values[0]
    df_orig["gdp"] = df_orig.gdp.values*deflator_2020_2005
    
    # Interpolate to yearly time step linearly
    df = df_orig.interpolate(method="polynomial",axis=0,order=2)
    
    # unit conversion
    df["emissions"] = df["emissions"]/1000.0 # mt co2/yr to gt co2/yr
    df["gdp"] = df["gdp"]/1000.0 # billion 2005 to trillion 2005
    df["population"] = df["population"]/1000.0 # million to billion
    
    # GDP growth
    df["gdp_growth"] = df.gdp.diff()/df.gdp.shift(periods=1)
    
    # kaya identity values
    df["e_g"] = df["energy"]/df["gdp"]
    df["f_e"] = df["emissions"]/df["energy"]
    df["g_p"] = df["gdp"]/df["population"]
    df["emissions_kaya"] = df.population * df.g_p * df.e_g * df.f_e
    
    return(df)
```

Main code for model

```python
def run_model(years,initial_gdp,initial_emissions,initial_dt,
              tcre,damage_coefs,pop_list,e_g_list,f_e_list,growth_list,
              adaptation_rate,mitigation_rate):
    
    # years over to which to run model
    # initial_gdp: world GDP in 2020 in trillions USD
    # initial_emissions: *cumulative* emissions in 2020 in GtCO2
    # initial_dt: temperature change relative to pre-industrial in 2020
    # tcre: transient climate response to cumulative emissions in degrees C per 1000 GtCO2 cumulative
    # damage_coefs: damage function coefficients, in a list: intercept, linear, quadratic
    # pop_list: exogenous population over the years desired
    # e_g_list: exogenous energy intensity of GDP 
    # f_e_list: exogenous emissions intensity of energy (but can be modified by mitigation)
    # growth_list: exogenous rate of economic growth
    # adaptation_rate: fraction reduction in the damage function per 100 trillion USD
    # mitigation_rate: fraction reduction in F/E per decade
    
    ## requires xarray, numpy, sys
        
    gdp_out = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    emissions_out = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    deltat_out = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    #f_e_out = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    
    gdp_out[0] = initial_gdp
    emissions_out[0] = initial_emissions
    deltat_out[0] = initial_dt
    
    # adaptation rate is fractional change in damage function per 100 trillion USD
    initial_adaptation = 0
    a = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    a[0] = initial_adaptation
    
    # mitigation rate 
    initial_mitigation = 0
    m = xr.DataArray(np.full(len(years),np.nan),coords=[years],dims=["time"])
    m[0] = initial_mitigation
    
    # damage function coefs
    # in percents -- 2% should be in damage_coefs as 2, not 0.02
    coef0 = damage_coefs[0]/100.0
    coef1 = damage_coefs[1]/100.0
    coef2 = damage_coefs[2]/100.0
    
    for i in np.arange(1,len(years),1):
        year = years[i]

        # exogenous vals 
        pop = pop_list[i]
        e_g = e_g_list[i]
        f_e = f_e_list[i]
        gr = growth_list[i]

        # iterate
        gdp_nodamage = gdp_out[i-1] + (gdp_out[i-1]*gr)

        # adaptation
        adapt = initial_adaptation + (((gdp_nodamage-initial_gdp)/100)*adaptation_rate)
        if adapt>1:
            adaptation = 1
        else:
            adaptation = adapt*1.0
        a[i] = adaptation
        
        # take cumulative emissions from the previous year as the tcre input to year t
        emissions_prev = emissions_out[i-1] # for input to tcre
        # calculate delta t
        deltat_out[i] = emissions_prev*(tcre/1000.0)
        #damages
        damages = coef2*(1-adaptation)*(deltat_out[i]**2) + (coef1*(1-adaptation)*deltat_out[i]) + (coef0*(1-adaptation))
        # gdp
        gdp_out[i] = gdp_nodamage - (gdp_nodamage*damages)
        
        # mitigation
        mitigat = initial_mitigation + ((year-np.amin(years))*mitigation_rate)
        #if mitigat>1:
        #    mitigation = 1
        #else:
        #    mitigation = mitigat*1.0
        f_e_new = f_e - mitigat #*(1-mitigation)
        if f_e_new<0:
            f_e_new = 0
        
        # emissions as a result of gdp
        emissions_out[i] = (pop * (gdp_out[i]/pop) * (e_g) * (f_e_new)) + emissions_out[i-1]
        
    return([gdp_out,emissions_out,deltat_out])
```

#### Analysis

Load Global Carbon Budget data so we have cumulative emissions from 1850

```python
gcp = pd.read_csv(loc_emissions+"GlobalCarbonBudget_HistoricalEmissions.csv")
gcp = gcp.loc[gcp.Year>=1850,:].reset_index().drop(columns="index")
gcp["total_emissions_gtc"] = gcp["fossil emissions excluding carbonation GtC"] + gcp["land-use change emissions GtC"]
gcp["total_emissions_gtco2"] = gcp.total_emissions_gtc*3.664
gcp["cumulative_emissions_gtco2"] = np.cumsum(gcp.total_emissions_gtco2.values)
gcp["obs_temp_change"] = gcp.cumulative_emissions_gtco2.values/1000.0 * tcre
```

Now load SSP data for desired scenario

```python
main_scenario = "SSP2"
scenario_subset = "Baseline" #Baseline, 19 for SSP1-19, 26 for SSP2-26, etc.
benchmark_mdls = {"SSP1":"IMAGE","SSP2":"MESSAGE-GLOBIOM",
                 "SSP3":"AIM/CGE","SSP4":"GCAM4","SSP5":"REMIND-MAGPIE"}
main_mdl = benchmark_mdls[main_scenario]
data_main = load_ssp_data(main_scenario,scenario_subset,main_mdl,np.arange(y1,y2+1,1))
```

Add cumulative emissions and temperature based on the GCP data

```python
data_main["cumulative_emissions"] = np.cumsum(data_main["emissions"]) + gcp.loc[gcp.Year==(y1-1),"cumulative_emissions_gtco2"].values
data_main["dt"] = data_main.cumulative_emissions/1000.0 * tcre
```

Mitigation rates from a variety of scenarios

```python
mitigation_scenarios = ["SSP1-19","SSP1-26","SSP2-19","SSP2-26","SSP2-34","SSP2-45","SSP3-34","SSP4-60"]
mitigation_scenario_rates = np.zeros(len(mitigation_scenarios))
for mmss in np.arange(0,len(mitigation_scenarios),1):
    ms = mitigation_scenarios[mmss]
    ms1 = ms.split("-")[0]
    ms2 = ms.split("-")[1]
    data_ms = load_ssp_data(ms1,ms2,benchmark_mdls[ms1],years)
    mitigation_scenario_rates[mmss] = np.polyfit(years,data_ms.f_e.values,1)[0]
    #print(np.polyfit(years,data_ms.f_e.values,1)[0])
```

Now simulate a range of adaptation and mitigation rates

```python
a_inputs = np.array([0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,100]) # stick a big number on the end for a "no climate damage" scenario
#m_inputs = np.array([0,,0.15])/100.0
m_inputs = np.linspace(0,0.0018,11)

gdp_adapt = xr.DataArray(np.full((len(years),len(a_inputs),len(m_inputs)),np.nan),
                          coords=[years,a_inputs,m_inputs],
                         dims=["time","adaptation","mitigation"])
emissions_adapt = xr.DataArray(np.full((len(years),len(a_inputs),len(m_inputs)),np.nan),
                          coords=[years,a_inputs,m_inputs],
                         dims=["time","adaptation","mitigation"])
temp_adapt = xr.DataArray(np.full((len(years),len(a_inputs),len(m_inputs)),np.nan),
                          coords=[years,a_inputs,m_inputs],
                         dims=["time","adaptation","mitigation"])
gdp_init = data_main.loc[data_main.year==y1,"gdp"].values[0]
emiss_init = data_main.loc[data_main.year==y1,"cumulative_emissions"].values[0]
t_init = data_main.loc[data_main.year==y1,"dt"].values[0]

for a in a_inputs:
    print(a)
    for m in m_inputs:
        gdp_adapt.loc[:,a,m], emissions_adapt.loc[:,a,m], temp_adapt.loc[:,a,m] = run_model(
                                            years,gdp_init,emiss_init,
                                            t_init,tcre,[damage_coef_0,damage_coef_1,damage_coef_2],
                                            data_main.population.values,data_main.e_g.values,
                                            data_main.f_e.values,data_main.gdp_growth.values,a,m)
        
# effect of adaptation on GDP, emissions, and temperature
temp_adapt_diff = temp_adapt[:,1:-1,:] - temp_adapt[:,0,:]
emissions_adapt_diff = emissions_adapt[:,1:-1,:] - emissions_adapt[:,0,:]
gdp_adapt_diff = gdp_adapt[:,1:-1,:] - gdp_adapt[:,0,:]
```

Run analysis for all SSPs

```python
scenarios = ["SSP1","SSP2","SSP3","SSP4","SSP5"]
m = 0
gdp_adapt_ssps = xr.DataArray(np.full((len(scenarios),len(years),len(a_inputs)),np.nan),
                              coords=[scenarios,years,a_inputs],
                                   dims=["scenario","time","adaptation"])
emissions_adapt_ssps = xr.DataArray(np.full((len(scenarios),len(years),len(a_inputs)),np.nan),
                              coords=[scenarios,years,a_inputs],
                                   dims=["scenario","time","adaptation"])
temp_adapt_ssps = xr.DataArray(np.full((len(scenarios),len(years),len(a_inputs)),np.nan),
                              coords=[scenarios,years,a_inputs],
                                   dims=["scenario","time","adaptation"])
gdp_main_ssps = xr.DataArray(np.full((len(scenarios),len(years)),np.nan),
                            coords=[scenarios,years],dims=["scenario","time"])
for s in scenarios:
    print(s)
    mdl = benchmark_mdls[s]
    data_s = load_ssp_data(s,"Baseline",mdl,years)
    data_s["cumulative_emissions"] = np.cumsum(data_s["emissions"]) + gcp.loc[gcp.Year==(y1-1),"cumulative_emissions_gtco2"].values
    data_s["dt"] = data_s.cumulative_emissions/1000.0 * tcre
    
    gdp_init = data_s.loc[data_s.year==y1,"gdp"].values[0]
    emiss_init = data_s.loc[data_s.year==y1,"cumulative_emissions"].values[0]
    t_init = data_s.loc[data_s.year==y1,"dt"].values[0]
    gdp_main_ssps.loc[s,:] = data_s.gdp.values
    
    for a in a_inputs:
        gdp_adapt_ssps.loc[s,:,a], emissions_adapt_ssps.loc[s,:,a], temp_adapt_ssps.loc[s,:,a] = run_model(
                                            years,gdp_init,emiss_init,
                                            t_init,tcre,[damage_coef_0,damage_coef_1,damage_coef_2],
                                            data_s.population.values,data_s.e_g.values,
                                            data_s.f_e.values,data_s.gdp_growth.values,a,m)
```

Combined effects of mitigation and adaptation

```python
fig = plt.figure(figsize=(8,8))
rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Helvetica Neue"]
rcParams["font.size"] = 20.0
rcParams["axes.linewidth"] = 1.3
rcParams['xtick.major.size'] = 8
rcParams['xtick.major.width'] = 1.3
rcParams['ytick.major.size'] = 8
rcParams['ytick.major.width'] = 1.3
rcParams["mathtext.default"] = "regular"

tfs = 21

gs1 = gridspec.GridSpec(1,1)
gs1.update(left=0.14,right=0.92,top=0.9,bottom=0.27,wspace=0.32,hspace=0.3)

ax = plt.subplot(gs1[0,0])

# colorbar
vmin = 0
vmax = 0.4
levels = np.arange(vmin,vmax+0.05,0.05)
ncol = 9
bounds = np.linspace(vmin,vmax,ncol)

cols = plt.get_cmap('Reds')(np.linspace(0,1,len(bounds)+1))
cm = colors.ListedColormap(cols[1:-1])
cm.set_over(cols[-1])
cm.set_under(cols[0])
norm = colors.Normalize(vmin=vmin,vmax=vmax)
cbar = plt.cm.ScalarMappable(norm=norm,cmap=cm)
ext = "both"

# heatmap

adapt_mesh, mitig_mesh = np.meshgrid(temp_adapt_diff.adaptation.values,
                                    temp_adapt_diff.mitigation.values*10*1000)

cs = ax.contourf(adapt_mesh,mitig_mesh,
                 temp_adapt_diff[-1,:,:].transpose("mitigation","adaptation"),
                 levels=levels,norm=norm,vmin=vmin,vmax=vmax,
            cmap=cm,extend=ext)
ax.contour(adapt_mesh,mitig_mesh,
                 temp_adapt_diff[-1,:,:].transpose("mitigation","adaptation"),
           colors="black",
           alpha=1,linewidths=1,linestyles="solid",levels=levels)
plt.xlabel("Fractional adaptation per \$100T")
plt.ylabel("Mitigation rate (GtCO$_2$ 1000EJ$^{-1}$ decade$^{-1}$)")

scenarios_plot = ["SSP1-26","SSP2-34","SSP2-45","SSP4-60"]
for s in scenarios_plot:
    yval = mitigation_scenario_rates[mitigation_scenarios.index(s)]*-10*1000
    plt.axhline(yval,color="gray",linewidth=1.5)
    s1 = s.split("-")[0]
    s2 = s.split("-")[1]
    lb = s1+"-"+s2[0]+"."+s2[1]
    plt.text(0.075,yval-(0.00025*1000),lb,fontsize=17,color="gray",
            backgroundcolor=cbar.to_rgba(0.01))

plt.title("Mitigation avoids warming penalty",fontsize=tfs,pad=10)
    
# plot colorbar
l = 0.14
b = 0.12
w = 0.78
h = 0.033
cbar_ax = fig.add_axes([l,b,w,h])
cb = fig.colorbar(cbar,cax=cbar_ax,orientation="horizontal",
                              drawedges=True,extend="max")
cbar_ax.tick_params(direction="in",width=0.05,labelsize=19)

cb.set_label("Warming due to adaptation in 2100 ($\degree$C)",fontsize=19)



plt.savefig("./Figures/Fig3.pdf")

plt.show()
```
